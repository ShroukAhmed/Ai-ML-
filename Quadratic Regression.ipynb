```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
 
x = np.arange(-5.0, 5.0, 0.1)
 
y = np.power(x, 2)
y_noise = 2 * np.random.normal(size = x.size)
ydata = y + y_noise #y=ax^2+bx+c where aâ‰ 0

plt.plot(x, ydata,  'go')
plt.plot(x, y, 'b')
plt.ylabel('Dependent Variable')
plt.xlabel('Independent Variable')
plt.show()
```
