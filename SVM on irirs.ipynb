#### The new library here is mlxtend, we used it in the lab and we were exposed to one of its functionalities like plot_decision_regions

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv("iris.csv")
spLength = np.array(iris.SepalLength.values)
ptLength = np.array(iris.PetalLength.values)
y = np.array(pd.factorize(iris.Name)[0].tolist())
X = np.column_stack((spLength, ptLength))
```

# C is the parameter that affects on the decision boundaries when changed. 
### 0 < C <= 1

## So, we were encourage that we are given this code without mlxtend and we tested the effect of C on the decision boundaries. Also what is the importance of the 'C' parameter.

```python
svm = SVC(C=0.9, kernel='linear')
svm.fit(X, y)
```

```python
plot_decision_regions(X, y, clf=svm, legend=4)

# Adding axes annotations
plt.xlabel('sepal length [cm]')
plt.ylabel('petal length [cm]')
plt.title('SVM on Iris')
plt.show()
```
